### Above scenario being done with try except block


```python
def divide2():
    try:
        a=int(input("Enter a number"))
        b=int(input("Enter another number"))
        c=a/b
        return c
    except Exception as e:
        print("Error Occured:",e)
        print("Please enter only numeric data greater than zero")

    
```

```python
d2_1=divide2()
```

```python
d2_2=divide2()
```

```python
def divide3():
    try:
        a = int(input("Enter a number"))
        b = int(input("Enter another number"))
        return a/b
    except ValueError as e:
        print("Please enter only Numeric data/Numbers")
```

```python
d3_1=divide3()
```

```python
d3_2=divide3()
```

```python
def divide4():
    try:
        a = int(input("Enter a number"))
        b = int(input("Enter another number"))
        return a/b
    except ValueError as e:
        print("Please enter only Numeric data/Numbers")
    except Exception as e:
        print("Please enter numbers only. DONOT Enter ZERO")
```

```python
d4_1=divide4()
```

```python
d4_2=divide4()
```

```python
def divide5():
    try:
        a = int(input("Enter a number"))
        b = int(input("Enter another number"))
        return a/b
    except ValueError as e:
        print("Please enter only Numeric data/Numbers")
    except ZeroDivisionError as e:
        print("DONOT Enter Zero. Enter only numbers other than zero")
    except Exception as e:
        print("Please enter any numeric data other than ZERO")
```

```python
d5_1=divide5()
```

```python
d5_2=divide5()
```

```python
d5_3=divide5()
```

### Finally block

```python
def divide6():
    try:
        a=int(input("Enter a number"))
        b=int(input("Enter another number"))
        return a/b
    except ValueError as e:
        print("Please enter only Numeric data")
    except ZeroDivisionError as e:
        print("DONOT Enter Zero.Enter only numbers other than zero")
    finally:
        print("This is a last statement")    
    
```

```python
d6_1=divide6()
```

```python
d6_1
```

```python
d6_2=divide6()
```
